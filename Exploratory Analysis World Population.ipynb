# Grouping By Continent
##### I grouped the data by continents to take a closer look at the growth rate per continent

```python
#Grouping my data by continent
df.groupby('Continent').mean('Continent')
```

```python
#Sorting by highest population in 2022
df.groupby('Continent').mean('Continent').sort_values(by='2022 Population', ascending = False)
```

```python
# Created new variable to trim out aspects of the original data set. I wanted only to see the Conntinent vs. Population.
df2 = df.groupby('Continent').mean('Continent').sort_values(by='2022 Population', ascending = False)
df2 = df2.drop(columns = ('Area (km²)'))
df2 = df2.drop(columns = ('Density (per km²)'))
df2 = df2.drop(columns = ('Growth Rate'))
df2 = df2.drop(columns = ('World Population Percentage'))
df2 = df2.drop(columns = ('Rank'))
df2
```

```python
#Refromating my x axis of my dataset so that the years are in ascending order
df3= df2.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2020 Population', '2022 Population']].mean('Continent')
df3
```

```python
df3 = df3.transpose()
df3
```

# Line Chart
#### Visual representation of the growth of each continent over the span of ~50 years
##### - Asia has been in the lead in population growth. 
##### - All other continents had population growth that is more stable over the course of 50 years.

```python
#Plotting Line Chart
df3.plot()
```

# Bar Chart of Population Growth Over the Decades
